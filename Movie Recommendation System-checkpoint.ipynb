# Movie Reccomedation System 

```python
import pandas as pd
```

```python
df=pd.read_csv('u.data',sep='\t',header=None)
df.head()
```

```python
df.drop([3],inplace=True,axis=1)
df.head()
```

```python
df.columns=['user_id','item_id','rating']
df.head()
```

```python
m_title=pd.read_csv('u.item',sep="\|",header=None)
m_title.head()
```

```python
m_title.drop([2,3,4],inplace=True,axis=1)
m_title
```

```python
genre_table=pd.DataFrame(m_title.loc[:,1:])
genre_table
```

```python
genre=pd.read_csv('u.genre',sep='\|',header=None)
genre.head()
```

```python
genre=genre[0]
genre
```

```python
genre_table.columns=['title']+list(genre)
```

```python
genre_table
```

```python
genre_table.set_index('title',inplace=True)
genre_table.head()
```

```python
genre_table=genre_table.T
genre_table
```

```python
genre_table.shape
```

```python
m_title=m_title[[0,1]]
```

```python
m_title.columns=['item_id','title']
```

```python
m_title
```

```python
df=pd.merge(df,m_title)
```

```python
df.head()
```

```python
df.shape
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
```

```python
ratings.head()
```

```python
ratings['Number of users']=df.groupby('title').count()['rating']
```

```python
ratings
```

```python
MovieRating=df.pivot_table(index='user_id',columns='title',values='rating')
MovieRating
```

## Main Function

```python
def Movie_Prediction(movie):
    Movie=MovieRating[movie]
    movie_by_rating=MovieRating.corrwith(Movie)
    Movie=genre_table[movie]
    movie_by_genre=genre_table.corrwith(Movie)
    
    corr_movies_rating=pd.DataFrame(movie_by_rating,columns=['Correlation'])
    corr_movies_rating.dropna(inplace=True)
    corr_movies_rating=corr_movies_rating.join(ratings['Number of users'])
    
    corr_movies_genre=pd.DataFrame(movie_by_genre,columns=['Correlation'])
    corr_movies_genre.dropna(inplace=True)
    
    Prediction_genre=corr_movies_genre.sort_values('Correlation',ascending=False)
    Prediction_rating=corr_movies_rating[corr_movies_rating['Number of users']>100].sort_values('Correlation',ascending=False)
    
    Prediction=pd.merge(Prediction_genre,Prediction_rating,on=['title'])
    
    print("Prediction by Genre")
    print(Prediction_genre)
    print("Prediction by Rating")
    print(Prediction_rating)
    print("Prediction")
    print(Prediction)
```

```python
Movie_Prediction('Hollow Reed (1996)')
```
